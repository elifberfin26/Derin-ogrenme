## Veriyi Tokenize Etme

Metin verilerini bir modele beslemeden önce, onları modelin anlayabileceği sayısal temsillerine dönüştürmemiz gerekir. Bu sürece "tokenization" denir. Burada, GPT-2 modeline uygun bir tokenizer kullanacağız.

```python
from datasets import Dataset

# Önce verimizi bir sözlük yapısına alıyoruz
train_data = {
    "text": [item["text"] for item in data]
}

# Bunu Hugging Face'in anlayacağı 'Dataset' formatına çeviriyoruz
dataset = Dataset.from_dict(train_data)

# Tokenizer işlemini tüm veriye uyguluyoruz
tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Dil modelleme için 'input_ids' aynı zamanda 'labels' olacaktır
tokenized_dataset = tokenized_dataset.map(lambda examples: {"labels": examples["input_ids"]}, batched=True)

print("Veri seti başarıyla hazırlandı ve tokenize edildi.")
display(tokenized_dataset)
```

```python
from transformers import GPT2Tokenizer

# GPT2'nin sözlüğünü kullanacağız
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Eğer metinler çok uzunsa sonuna boşluk ekle (padding)
tokenizer.pad_token = tokenizer.eos_token

# Veriyi modele uygun hale getiriyoruz
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=32)

print("Tokenize fonksiyonu ve GPT2 tokenizer başarıyla tanımlandı.")
```

## Dil Modelini Yükleme

GPT-2 gibi önceden eğitilmiş bir dil modeli, metin üretimi, çeviri veya metin tamamlama gibi birçok doğal dil işleme görevi için kullanılabilir. Burada, Hugging Face `transformers` kütüphanesinden `GPT2LMHeadModel`'i yüklüyoruz.

```python
from transformers import GPT2LMHeadModel

# Hazır bir dil modelini (GPT2) yükle
model = GPT2LMHeadModel.from_pretrained("gpt2")

print("GPT2 dil modeli başarıyla yüklendi.")
```

## Veri Setini Hazırlama ve Tokenize Etme

Modeli eğitebilmek veya çıkarım yapabilmek için verimizi Hugging Face `datasets` kütüphphanesinin `Dataset` formatına dönüştürmemiz ve ardından tokenization işlemini uygulamamız gerekiyor.

```python
from datasets import Dataset

# Önce verimizi bir sözlük yapısına alıyoruz
train_data = {
    "text": [item["text"] for item in data]
}

# Bunu Hugging Face'in anlayacağı 'Dataset' formatına çeviriyoruz
dataset = Dataset.from_dict(train_data)

# Tokenizer işlemini tüm veriye uyguluyoruz
tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Dil modelleme için 'input_ids' aynı zamanda 'labels' olacaktır
tokenized_dataset = tokenized_dataset.map(lambda examples: {"labels": examples["input_ids"]}, batched=True)

print("Veri seti başarıyla hazırlandı ve tokenize edildi.")
display(tokenized_dataset)
```

## Modeli Eğitme

Veri setimiz hazır olduğuna göre, şimdi modelimizi eğitmeye başlayabiliriz. Hugging Face `Trainer` sınıfı, model eğitimini kolaylaştıran birçok kullanışlı özellik sunar.

```python
from transformers import TrainingArguments, Trainer, DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

training_args = TrainingArguments(
    output_dir="./moda_modeli",          # Modelin kaydedileceği yer
    num_train_epochs=3,                  # Tüm veriyi kaç kere okuyacak (3 idealdir)
    per_device_train_batch_size=2,       # Her seferinde kaç örnek görecek
    save_steps=10,                       # Kaç adımda bir yedek alacak
    logging_steps=1,                     # Gelişimi ne sıklıkla ekrana yazdıracak
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
)

print("Trainer nesnesi başarıyla oluşturuldu.")
```

```python
print(training_args)
```

```python
trainer.train()

print("Model eğitimi tamamlandı.")
```

## Modeli Test Etme

Eğittiğimiz modelin nasıl çalıştığını görmek için bir test fonksiyonu oluşturalım. Bu fonksiyon, belirli bir duygu ve hava durumuna göre kombin önerisi üretecek.

```python
# Modeli test etmek için bir fonksiyon yazalım
def kombin_onerisi_yap(duygu, hava):
    # Senin belirlediğin formata uygun bir girdi hazırlıyoruz
    input_text = f"Bağlam: {duygu}, {hava}. Öneri:"

    # Metni sayıya (token) çeviriyoruz
    inputs = tokenizer(input_text, return_tensors="pt")

    # Modelin pad_token_id'sini tokenizer'ın eos_token_id'si olarak ayarlıyoruz
    # Bu, GPT2 gibi modeller için metin üretiminde yaygın bir ayarlamadır.
    model.config.pad_token_id = tokenizer.eos_token_id

    # Modelden yeni kelimeler üretmesini istiyoruz
    # Senin planındaki 'Temperature' (yaratıcılık) parametresini burada kullanıyoruz
    outputs = model.generate(
        inputs["input_ids"],
        attention_mask=inputs["attention_mask"], # Dikkat maskesini modele iletiyoruz
        max_length=50,
        num_return_sequences=1,
        temperature=0.7,  # Bu değer arttıkça model daha 'çılgın' öneriler yapar
        do_sample=True    # Rastgelelik ekleyerek her seferinde aynı şeyi dememesini sağlar
    )

    # Sayıları tekrar okunabilir metne çeviriyoruz
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# ŞİMDİ TEST EDELİM!
sonuc = kombin_onerisi_yap("Enerjik", "Karlı")
print(sonuc)
```
